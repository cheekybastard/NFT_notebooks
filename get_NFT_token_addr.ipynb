```python
import httpx
# https://devhints.io/xpath
import lxml.html
from cytoolz import curried as c
import cytoolz as tz

tokens = tz.pipe(
    range(2),
    c.map(lambda x: lxml.html.fromstring(httpx.get(f'https://polygonscan.com/tokens-nft?ps=100&p={x}').content)),
    c.map(lambda x: x.xpath('//div[@class="token-wrap"]/a/@href')),
    c.concat,
    c.map(lambda x: x.split('/token/')[1]),
    list
)
print(len(tokens))
assert all(42 == x for x in list(map(len, tokens)))
tokens
```
